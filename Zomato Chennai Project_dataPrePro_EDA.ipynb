## 4. Data Preprocessing

```python
raw_data.replace(to_replace = ['None','Does not offer Dining','Not enough Delivery Reviews','Does not offer Delivery',
                                'Not enough Dining Reviews'],value = '0', inplace = True)
raw_data.head()
```

```python
raw_data.isnull().sum()
```

```python
raw_data.dtypes
```

```python
# Changing the datatypes
raw_data['Dining Rating'] = raw_data['Dining Rating'].astype('float64')
raw_data['Dining Rating Count'] = raw_data['Dining Rating Count'].astype('int64')
raw_data['Delivery Rating'] = raw_data['Delivery Rating'].astype('float64')
raw_data['Delivery Rating Count'] = raw_data['Delivery Rating Count'].astype('int64')
```

```python
raw_data.info()
```

### Working with 'Cuisine', 'Top Dishes' & 'Features' columns

```python
# Cuisine
raw_data['Cuisine'] = raw_data['Cuisine'].str.replace('[','')
raw_data.head(5)
raw_data['Cuisine'] = raw_data['Cuisine'].str.replace(']','')
raw_data.head(5)
raw_data['Cuisine'] = raw_data['Cuisine'].str.replace("'",'')
raw_data.head(5)

# Top Dishes
raw_data['Top Dishes'] = raw_data['Top Dishes'].str.replace('[','')
raw_data.head(5)
raw_data['Top Dishes'] = raw_data['Top Dishes'].str.replace(']','')
raw_data.head(5)
raw_data['Top Dishes'] = raw_data['Top Dishes'].str.replace("'",'')
raw_data.head(5)

# Features
raw_data['Features'] = raw_data['Features'].str.replace('[','')
raw_data.head(5)
raw_data['Features'] = raw_data['Features'].str.replace(']','')
raw_data.head(5)
raw_data['Features'] = raw_data['Features'].str.replace("'",'')
raw_data.head(5)
```

```python
# Deleting unwanted column
raw_data.drop(['Zomato URL','Address'],axis = 1, inplace = True)
```

## 5. Exploratory Data Analysis

#### Q1)  What are the top locations in Chennai with the highest number of restaurants?

```python
top_locations_df = zomato_df.groupby(['Location']).agg({'Name of Restaurant':'count'}).rename(columns={'Name of Restaurant' : 'No of Restaurants'})
top_locations_df = top_locations_df.sort_values(by=['No of Restaurants'],ascending = False).reset_index()
top_locations_df.head(5)
```

```python
fig = px.scatter(top_locations_df, x = 'No of Restaurants', y = 'Location', color = 'No of Restaurants',
                  title = 'Top locations with Most Restaurants')
fig.show()
```

#### Q2) What is the distribution of restaurant prices (Price for 2) across different locations?

```python
location_rest_price_df = zomato_df.groupby(['Location'])['Price for 2'].mean().reset_index()
location_rest_price_df = location_rest_price_df.rename(columns={'Price for 2' : 'Average Price'})
location_rest_price_df
```

```python
fig = px.scatter(location_rest_price_df, x = 'Average Price', y = 'Location', color = 'Average Price',
                  title = 'Top locations with Most Restaurants').update_traces(marker={ "size" : 5, })
fig.show()
```

#### Q3) Which cuisines are the most popular in Chennai based on restaurant listings?

```python
popular_cuisine_df = zomato_df['Cuisine'].value_counts().reset_index()
popular_cuisine_df.rename(columns={'count':'Counts of Restaurant'}, inplace=True)
popular_cuisine_df
```

#### Q4) How does the average price for two vary across different types of cuisines?

```python
cuisine_rest_price_df = zomato_df.groupby(['Cuisine'])['Price for 2'].mean().reset_index()
cuisine_rest_price_df = cuisine_rest_price_df.rename(columns={'Price for 2':'Average Price'})
cuisine_rest_price_df
```

#### Q5) What are the most common features offered by restaurants (e.g., home delivery, table booking)?

```python
most_features_offered_df = zomato_df['Features'].value_counts().reset_index()
most_features_offered_df.rename(columns={'count':'Counts of Restaurant'}, inplace=True)
most_features_offered_df
```

#### Q6) What is the distribution of dining and delivery ratings across restaurants?

```python
dining_delivery_dist_df = zomato_df.groupby('Name of Restaurant').agg({'Dining Rating':'mean','Delivery Rating':'mean'}).reset_index()
dining_delivery_dist_df.sort_values(by=['Dining Rating','Delivery Rating'],ascending=False,inplace =True)
dining_delivery_dist_df 
```

### Q7) Is there a correlation between dining ratings and delivery ratings?

```python
Corr_dining_delivery_df = zomato_df['Dining Rating'].corr(zomato_df['Dining Rating'])
Corr_dining_delivery_df
```

```python
correlation_matrix = zomato_df[['Dining Rating', 'Delivery Rating']].corr()
correlation_matrix
```

#### Answer : There is a strong relationship between Dining Ratings and Delivery Ratings.

#### Q8) Do restaurants with higher dining ratings also have higher delivery ratings?

```python
DiningRating_DeliveryRating_df = zomato_df[['Dining Rating','Delivery Rating']].corr()
DiningRating_DeliveryRating_df
```

#### Answer : Yes, higher Dining Ratings also have higher Delivery Ratings. It is Positively correlated.
- The correlation value is a number between -1 and 1:
- A value closer to 1 means a strong positive relationship (higher dining ratings are associated with higher delivery ratings).
- A value closer to -1 means a strong negative relationship (higher dining ratings are associated with lower delivery ratings).
- A value close to 0 indicates little to no linear relationship between the two ratings.

#### Q11) Which regions or neighborhoods specialize in specific types of cuisine (e.g., Biryani, Chinese)?

```python
zomato_df.head()
```

```python
# Step 1: Split the 'Cuisine Type' column into individual cuisines  

zomato_df['Cuisine'] = zomato_df['Cuisine'].str.split(',')
```

```python
# Step 2: Explode the list of cuisines into individual rows

exploded_df = zomato_df.explode('Cuisine')
exploded_df
```

```python
# Step 3: Group by 'Region' and 'Cuisine Type' and count the number of restaurants
cuisine_by_region = exploded_df.groupby(['Location','Cuisine']).size().reset_index(name='Restaurant Count')
cuisine_by_region
```

```python
# Step 4: Find the most common cuisine in each region

dominant_cuisine_by_region = cuisine_by_region.groupby('Location').first().reset_index()
dominant_cuisine_by_region.head(20)
```

```python
fig = px.histogram(dominant_cuisine_by_region, x = 'Restaurant Count', y = 'Location', color = 'Cuisine',
                  title = 'Dominant Cuisine By Location')
fig.show()
```
